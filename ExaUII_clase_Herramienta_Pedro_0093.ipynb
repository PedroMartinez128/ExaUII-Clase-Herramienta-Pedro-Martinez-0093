```python
class Herramienta:
    def __init__(self, nombre, tipo, marca):
        self.nombre = nombre
        self.tipo = tipo
        self.marca = marca

    def leerDatos(self):
        print("--- Registro de Herramienta ---")
        self.nombre = input("Ingrese el nombre de la herramienta: ")
        self.tipo = input("Ingrese el tipo (Manual, Eléctrica, etc.): ")
        self.marca = input("Ingrese la marca: ")

    def mostrarDatos(self):
        print("\n--- Detalles de la Herramienta ---")
        print("Nombre :", self.nombre)
        print("Tipo   :", self.tipo)
        print("Marca  :", self.marca)

# --- Creación de los objetos herramienta1 y herramienta2 ---

print("=======================================")
print("      INVENTARIO DE HERRAMIENTAS       ")
print("=======================================")

# Objeto 1
print("\n--- Herramienta Numero 1 ---")
herramienta1 = Herramienta("", "", "") # Creando el objeto herramienta1
herramienta1.leerDatos()
herramienta1.mostrarDatos()

print("---------------------------------------")

# Objeto 2
print("\n--- Herramienta Numero 2 ---")
herramienta2 = Herramienta("", "", "") # Creando el objeto herramienta2
herramienta2.leerDatos()
herramienta2.mostrarDatos()

print("=======================================")
print("Programa finalizado =)")
print("Software desarrollado Por Pedro Martinez NL = 34")
```
